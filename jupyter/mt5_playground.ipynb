```python
import MetaTrader5 as mt5
```

```python
mt5.initialize()
```

```python
candles_m1 = mt5.copy_rates_from_pos("Step Index", mt5.TIMEFRAME_M1, 0, 100)
candles_m15 = mt5.copy_rates_from_pos("Step Index", mt5.TIMEFRAME_M15, 0, 100 * 16)
candles_h4 = mt5.copy_rates_from_pos("Step Index", mt5.TIMEFRAME_H4, 0, 100)
```

```python
import numpy as np
```

```python
np.savetxt("step_index_m1.csv", candles_m1, delimiter=",", fmt="%d,%f,%f,%f,%f,%d,%d,%d")
np.savetxt("step_index_m15.csv", candles_m15, delimiter=",", fmt="%d,%f,%f,%f,%f,%d,%d,%d")
np.savetxt("step_index_h4.csv", candles_h4, delimiter=",", fmt="%d,%f,%f,%f,%f,%d,%d,%d")
```
